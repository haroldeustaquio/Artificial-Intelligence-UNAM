### 2 Sensores:
- Ubicacion (extender a N cuartos seguidos)
- Status: “Sucio”/”Limpio”

### 3 Acciones:
- Aspirar
- Mover a la izquierda
- Mover a la derecha

image.png


```python
class Environment(object):
    def __init__(self,initial): # [True,True]
        self.status = initial
        self.legend = {False:"Limpio",True:"Sucio"}

    def __str__(self):
        output = "Estado de la casa\n--------------------------------\n"
        output = output + str([f"{self.legend[room]}" for room in self.status])
        return output

    def get_environment(self,room):
        return(self.status[room])

    def set_environment(self,room):
        self.status[room] = False
```

```python
class ReflexAgent(object):
    def __init__(self,room):
        self.room = room
        self.step = +1

    def __str__(self):
        return f"Estoy en el Cuarto {self.room}"

    def precept_and_act(self,env):
        # If Sucio -> Limpiar
        n = len(env.status)
        if env.get_environment(self.room):
            env.set_environment(self.room)
        else:
            if self.room + self.step == n:
                self.step = -1
            elif self.room + self.step == -1:
                self.step = +1
            self.room = self.room + self.step
        print(self)
        print(env)

```

```python
house = Environment([True, True, True, False, True])
print(house)

vacuum = ReflexAgent(0)
print(vacuum)
```

```python
for time in range(10):
    print(time)
    vacuum.precept_and_act(house)
```
